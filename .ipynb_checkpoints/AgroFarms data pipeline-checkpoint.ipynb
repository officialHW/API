#### Retrieving Weather Data for Multiple Cities 

###### By iterating over multiple cities, I can analyze and compare weather conditions across different locations.

###### 1. I will define a list of cities for which I want to retrieve weather data.
###### 2. Using a loop, I will iterate over each city in the list and construct the API URL with the city name as a parameter.
###### 3. I will send an HTTP GET request to the OpenWeatherMap API endpoint for each city and parse the JSON response into a Python dictionary.
###### 4. From the response, I extract relevant weather parameters such as city ID, name, temperature, weather condition, and date and timezone.

```python
## get input locations
all_locations = pd.read_excel('cities.xlsx')
all_locations.head()
```

```python
## Retrieve weatheer data for all the cities of interest

# creating an empty list to hold data points where we will keep the below:
Locations = []
Weathers = []
Descriptions = []
Temperatures = []
Pressures = []
Humidities = []
Wind_speeds = []
Dates = []
Timezones = []

api_key = '8499fd113edeba571098b8402fb571f1'

for index, row in all_locations.iterrows():
    lat = row['latitude'] # here we are saying for each row the latitude for them is what should be inputted here e.g for abeokuta put its latitude which you can see on its row
    lon = row['longitude'] # here we are saying for each row the lontitude for them is what should be inputted here  e.g for abeokuta put its lontitude which you can see on its row
    
    url = f'https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}'
    
    response = requests.get(url)
    data = response.json() # here we pass the api to a json to be able to extract

    # step 4: extracting relevant data for all locations of interest
    location = data['name']
    Locations.append(location) # here i am telling python that for every location name it gets it should append or put it into the locations list, same applies to the rest as shown below to keep it safe

    weather = data['weather'][0]['main']
    Weathers.append(weather)

    description = data['weather'][0]['description']
    Descriptions.append(description)

    temperature = data['main']['temp']
    Temperatures.append(temperature)

    pressure = data['main']['pressure']
    Pressures.append(pressure)

    humidity = data['main']['humidity']
    Humidities.append(humidity)

    wind_speed = data['wind']['speed']
    Wind_speeds.append(wind_speed)

    dt = data['dt']
    Dates.append(dt)

    timezone = data['timezone']
    Timezones.append(timezone)




```

## Initial Transformation

```python
# Creating my dataframe to make it more organized
df = pd.DataFrame({'Location': Locations,
                    'Weather': Weathers,
                    'Description': Descriptions,
                    'Temperature': Temperatures,
                    'Pressure': Pressures,
                    'Humidity':Humidities,
                    'Wind_speed': Wind_speeds,
                    'dt': Dates,
                    'Timezone': Timezones})

df
```

```python
# making a copy of the dataframe for further transformation purpose
df2 = df.copy()
```

## TRANSFORMATIONS TO THE EXTRACTED DATA 

```python
## convert the temperature readings to celcius

# for this if the api gives us a simple way to do this we could follow it, or we could go another route of converting the temperature to celcius as show in the below cell using a lambda function

# applying it to the column temperature
df2['Temperature'] = df2['Temperature'].apply(lambda x: x - 273.15)
```

```python
# cofirming the temperature has been converted to celcius
df2
```

```python
# converting the date to a readable format
#Â i want the function which i will be defining to take not just the date but also the timezone, 
def adjust_date(row): # i want the function to take a row
    unix_datetime = row['dt'] + row['Timezone']     # and here i want the unit_datetime to take a submission of the date and the timezone so as to get the actual time

    # conversion to datetime itself
    actual_datetime = datetime.fromtimestamp(unix_datetime)

    return actual_datetime
```

```python
# since i need just the day because i they want to be viewing this once a day hence would remove the time
df2.apply(adjust_date, axis = 1)

df2['reading_date'] = df2.apply(adjust_date, axis = 1) # here i am applying to the row, telling python i want it to apply it on the column row-wise
```

```python
# check result
df2
```

```python
# now I will drop the dt and Timezone columns
df2.drop(['dt', 'Timezone'], axis=1, inplace=True)
```

```python

from sqlalchemy import create_engine

from dotenv import load_dotenv
import os

load_dotenv()
```

```python
from snowflake.sqlalchemy import URL
```

```python
engine = create_engine (URL(
        user = os.getenv('user'),
        password = os.getenv('password'),
        account = os.getenv('account_identifier'),
        database = os.getenv('database'),
        schema = os.getenv('schema'),
        warehouse = os.getenv('warehouse')
))

```

```python
df2.to_sql('weather', con=engine, index=False, if_exists='append')
```
